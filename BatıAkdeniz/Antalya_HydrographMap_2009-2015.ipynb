```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
agi=pd.read_excel("Agi_ky_2006-2015.xlsx")
dis24=pd.read_excel('Dis24_ky_2006-2015.xlsx')
chosen=pd.read_excel("Chosen_all.xlsx")
```

```python
#Define coordinates of where we want to center our map
boulder_coords = [36.884804, 30.704044]
```

```python
my_map = folium.Map(location = boulder_coords, zoom_start = 8,control_scale=True)
```

```python
my_map
```

```python
east = chosen.east
north = chosen.north
name = chosen.stname
elevation = chosen.elev
area = chosen.area
# Variables from chosen file
```

```python
east.head(3)
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

```python
for i in range(len(north)):
    names = name.iloc[i]
    elv = elevation.iloc[i]
    are = area.iloc[i]
    html=f"<div><p>Staion Name:{names}</p></div><b><img src=https://raw.githubusercontent.com/Talenda/Antalya_2006_2015/main/pltm/{names}.png?h=10%&w=10%></b><div><p>Elevation:{elv} meters</p><div><div><p>Area:{are}km²</p><div>"
    iframe = folium.IFrame(html)
    popup = folium.Popup(iframe,
                     min_width=700,
                     max_width=700)
    marker = folium.Marker(
       location=[north.iloc[i],east.iloc[i]],
       tooltip= name.iloc[i],
       icon=folium.Icon(icon="cloud", color ='darkblue'),
       popup= popup
)
    marker.add_to(my_map)
```

```python
my_map
```

# Layer Control with map types

```python
# show map types using layer control

# map
map_layer_control = folium.Map(location=boulder_coords, zoom_start=8)

# add tiles to map
folium.raster_layers.TileLayer('Open Street Map').add_to(my_map)
folium.raster_layers.TileLayer('Stamen Terrain').add_to(my_map)
folium.raster_layers.TileLayer('Stamen Toner').add_to(my_map)
folium.raster_layers.TileLayer('Stamen Watercolor').add_to(my_map)
folium.raster_layers.TileLayer('CartoDB Positron').add_to(my_map)
folium.raster_layers.TileLayer('CartoDB Dark_Matter').add_to(my_map)

# add layer control to show different maps
folium.LayerControl().add_to(my_map)

# display map
my_map
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

# Plugins (minimap, zoom toggler, full screen)

```python
# mini map, scroll zoom toggle button, full screen

# plugin for mini map
minimap = plugins.MiniMap(toggle_display=True)

# add minimap to map
my_map.add_child(minimap)

# add scroll zoom toggler to map
#plugins.ScrollZoomToggler().add_to(my_map)

# add full screen button to map
plugins.Fullscreen(position='topright').add_to(my_map)

# display map
my_map
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

# Measure Plugin

```python
# measure control
measure_control = plugins.MeasureControl(position='topleft', 
                                         active_color='red', 
                                         completed_color='red', 
                                         primary_length_unit='kilometers')
# add measure control to map
my_map.add_child(measure_control)
# display map
my_map
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

# Draw Plugin

```python
# draw tools
# export=True exports the drawn shapes as a geojson file
draw = plugins.Draw(export=True)

# add draw tools to map
draw.add_to(my_map)

# display map
my_map
```
